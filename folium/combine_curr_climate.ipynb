```python
import pandas as pd

# Open the climate TAWES stations
curr = pd.read_csv('C:\\Users\\sh16450\\OneDrive - University of Bristol\\Documents\\GitHub\\openweatherviz\\folium\\current_csv\\2024-11-01_21-30-00-00_tawes_current_10min.csv')
# Open the current TAWES readings
klima = pd.read_csv('C:\\Users\\sh16450\\OneDrive - University of Bristol\\Documents\\GitHub\\openweatherviz\\folium\\20241101-212000_10min_TAWES.csv')

# Open the station info for the current data and merge
curr_station_info = pd.read_csv('C:\\Users\\sh16450\\OneDrive - University of Bristol\\Documents\\GitHub\\openweatherviz\\folium\\current_csv\\'
                                + 'corrected_station_info_current_tawes.csv').rename(columns={'id':'station'})
merged_curr = pd.merge(curr, curr_station_info, on='station')

# Open the station info for the climate data and merge
station_data = pd.read_csv('C:\\Users\\sh16450\\OneDrive - University of Bristol\\Documents\\GitHub\\openweatherviz\\folium\\station_info_new.csv').rename(columns={'id':'station'})
merged_data = pd.merge(klima, station_data, on='station')

# Rename current data so it fits with the climate data
new_curr = merged_curr.rename(columns={'TL':'tl', 'TP':'tp', 'TLMAX':'tlmax', 'TLMIN':'tlmin', 
                                       'TS':'ts','FF':'ff', 'FFX':'ffx','DD':'dd', 'DDX': 'ddx', 'GLOW':'cglo',
                                       'P':'p', 'RF':'rf', 'RR':'rr', 'SCHNEE':'sh', 'SO':'so'})

new_curr = new_curr.drop(['tp', 'PRED', 'RRM', 'ff'], axis=1)

# Merge climate and current data
combined = pd.concat([merged_data, new_curr])
combined.to_csv('merged_data.csv')
```

```python
# Drop rows that have too man NANs, then drop duplicates based on a subset of values, and then sort the values by name and time
extract = combined.dropna(thresh=combined.shape[1]-4, axis=0).drop_duplicates(subset=['tl', 'tlmax', 'rf', 'dd']).sort_values(by=['name', 'time'])
extract
extract.loc[extract.name == 'Melk']
```
